```python
###########
# Imports #
###########

from ultralytics import YOLO
from ultralytics.solutions import object_counter
import cv2
import numpy as np
```

```python
######################
# Model & parameters #
######################


## File used to track ##
filename = "Production_Line_2"

## Initialize segmentation model ##
model = YOLO("yolov8n-seg.pt")

## Create VideoCapture object ##
cap = cv2.VideoCapture(f"/home/valerio/code/ValerioCann/Mod√®les/Data/Videos//{filename}.mp4")

## Retrieve height, widht & frame per second ##
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

## Define the region points ##
region_points = [(800, 70), (1104, 70), (1104, 1035), (800, 1035)]
#line_points = [(400, 200), (400, 1080)] 

## Class to track ##
classes_to_count = [39]  #'bottle'


###################################
# Initialize objects for tracking #
###################################


## Create the VideoWriter object ##
video_writer = cv2.VideoWriter("object_counting_output.avi",
                       cv2.VideoWriter_fourcc(*'mp4v'),
                       fps,
                       (w, h))

## Initialize the Object Counter ##
counter = object_counter.ObjectCounter()
counter.set_args(view_img=True,
                 reg_pts=region_points,
                 classes_names=model.names,
                 view_out_counts=False,
                 draw_tracks=False,
                 count_reg_color=(255,255, 100),
                 count_color=(255,255, 100),
                 region_thickness=2,
                 line_dist_thresh=5)


#######################
# Tracking & Counting #
#######################


while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    tracks = model.track(im0, persist=True, show=False,
                         classes=classes_to_count)

    im0 = counter.start_counting(im0, tracks)
    video_writer.write(im0)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
